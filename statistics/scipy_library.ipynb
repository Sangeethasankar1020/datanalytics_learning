```python
mu,sigma=0,0.1 
```

```python
s=np.random.normal(mu,sigma,1000) #generating 1000 samples
```

**Overlay the Probability Density Function - PDF**

To overlay the PDF of the normal distribution, you first need to create a range of x values and then compute the corresponding y values using the PDF formula.

```python
#plt.xlim(): Retrieves the current limits of the x-axis to ensure the PDF aligns with the histogram.
#np.linspace(): Generates 100 evenly spaced values between xmin and xmax.
#stats.norm.pdf(): Computes the probability density function for each value in x.
#plt.plot(): Plots the PDF line over the histogram.
```

```python
xmin,xmax=plt.xlim()
x=np.linspace(xmin,xmax,100)
p=stats.norm.pdf(x,mu,sigma)
plt.plot(x,p,'k',linewidth=1)
plt.title('Normal Distribution')
plt.show()
```

### Optimization

```python
#simple function to minimize
def objective_function(x):
    return x**2 +5 *np.sin(x)

result=minimize(objective_function,x0=0) #start at 0

print("Optimized Value:", result.x)
print("Objective Function Value:", result.fun)
```

### Interpolation 

SciPy is useful for interpolation, which is essential when working with datasets that require smooth transitions between known data points.

```python
#data points
```

```python
x=np.array([0,1,2,3,4,5])
y=np.array([0,1,0,1,0,1])
```

```python
#interpolation function
f=interp1d(x,y,kind='cubic')


#new x values for interpolation
x_new=np.linspace(0,5,num=50)
y_new=f(x_new)

plt.plot(x, y, 'o', label='Known Data Points')
plt.plot(x_new, y_new, '-', label='Interpolated Function')
plt.title('Interpolation Example')
plt.legend()
plt.show()
```

### Integrating Functions

```python
#integarte function

def f(x):
    return x**2

result,error=quad(f,0,1) #0-1

print("Integral result:", result)
print("Estimated error:", error)
```
