### Plot Customization 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
```

```python
fig , axes = plt.subplots()
axes.plot(x,x**2,"r.-")
axes.plot(x,x**3,"b--")
```

```python
fig , axes = plt.subplots()
axes.plot(x,x**2,color="#e600e6",alpha=0.5)
axes.plot(x,x**3,color="#a31aff")
```

```python
fig , axes = plt.subplots()
axes.plot(x,x**2,color="#e600e6",lw=3,ls="--")
```

```python
fig , axes = plt.subplots()
# axes.plot(x,x**2,color="#e600e6",lw=3,ls="--",marker="o")
# axes.plot(x,x**2,color="#e600e6",lw=3,ls="--",marker="s")
# axes.plot(x,x**2,color="#e600e6",lw=3,ls="--",marker="+")
axes.plot(x,x**2,color="orange",lw=3,ls="--",marker="o",markersize=10,markerfacecolor="green",markeredgecolor="red",markeredgewidth=2)
```

### Plot Range

```python
fig,axes=plt.subplots(1,3,figsize=(10,5))


axes[0].plot(x,x**2,x**3)
axes[0].set_title("Axes Zero")

axes[1].plot(x**2,x**3,x**4)
axes[1].set_title("Axes One")

axes[2].plot(x,x**6,x**9)
axes[2].set_title("Axes Two")

```
