```python
import pandas as pd
```

```python
#reshaping
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [90, 85, 80],
    'Science': [88, 92, 85],
    'English': [95, 85, 89]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
melted=df.melt(id_vars="Name",var_name="Subject",value_name="Grade")
```

```python
melted
```

```python
stacked=df.set_index("Name").stack()
```

```python
stacked
```

```python
unstacked=stacked.unstack()
```

```python
unstacked
```

```python
data = {
    'OrderID': [1, 2, 3, 4, 5, 6],
    'Product': ['Shoes', 'Shoes', 'Shirt', 'Shirt', 'Shirt', 'Shoes'],
    'Category': ['Footwear', 'Footwear', 'Apparel', 'Apparel', 'Apparel', 'Footwear'],
    'Quantity': [2, 3, 1, 5, 2, 4],
    'Price': [50, 50, 30, 30, 30, 50],
    'City': ['New York', 'Los Angeles', 'New York', 'New York', 'Los Angeles', 'New York']
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df["Total_Revenue"]=df["Quantity"]*df["Price"]
```

```python
df
```

```python
pivot_table=df.pivot_table(index="City",columns="Product",values="Total_Revenue",aggfunc="sum",margins=True)
```

```python
pivot_table
```

```python
#merging
```

```python
df1=pd.DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']},index=[0,1,2,3]
)
```

```python
df2=pd.DataFrame({
    'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']},index=[4,5,6,7]
)
```

```python
df3=pd.DataFrame({
    'A':['A8','A9','A10','A11'],
    'B':['B8','B9','B10','B11'],
    'C':['C8','C9','C10','C11'],
    'D':['D8','D9','D10','D11']},index=[8,9,10,11]
)
```

```python
df=pd.concat([df1,df2,df3])
```

```python
df
```

```python
df=pd.concat([df1,df2,df3],axis=1)
```
