```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
#get year and month for analyse
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
```

### EDA

**sales distribution**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], kde=True, bins=30)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

**Business Insight:**  identify the frequency of small vs. large sales. If most sales are small,it  indicate that the **business depends on high sales volume**  than a few large orders.

**sales category**

```python
#which category sales most
```

```python
category_sales = df.groupby('Category')['Sales'].sum().reset_index()
```

```python
category_sales
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Sales', data=category_sales)
plt.title('Total Sales by Category')
plt.show()
```

**Business Insight:**  categories **Furniture, Office Supplies, or Technology** contribute the most to total sales.  company can focus on most

**Region sales**

```python
region_sales = df.groupby('Region')['Sales'].sum().reset_index()
```

```python
region_sales
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Sales', data=region_sales)
plt.title('Total Sales by Region')
plt.show()
```

**Business Insight:**  it suggest focusing sales efforts on high-performing regions

**sales over time** (year)

```python
yearly_sales = df.groupby('Order Year')['Sales'].sum().reset_index()
```

```python
yearly_sales
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x='Order Year', y='Sales', data=yearly_sales, marker='o')
plt.title('Sales Over Time (Yearly)')
plt.show()
```

**Business Insight:** identify periods of growth or decline and inform long-term planning.

**by month**

```python
monthly_sales = df.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()
```

```python
monthly_sales
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Order Month', y='Sales', hue='Order Year', data=monthly_sales, marker='o')
plt.title('Sales Trends by Year and Month')
plt.show()
```

**customer segment analysis**

```python
segment_sales = df.groupby('Segment')['Sales'].sum().reset_index()
```

```python
segment_sales
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Segment', y='Sales', data=segment_sales)
plt.title('Total Sales by Customer Segment')
plt.show()
```

**Business insights:** we can focus on general consumer they puchased most.
