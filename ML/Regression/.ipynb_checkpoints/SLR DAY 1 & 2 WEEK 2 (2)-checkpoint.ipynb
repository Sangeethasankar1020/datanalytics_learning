# Assumption - 2 Normality of residuals

```python
xc = xc.drop(['age','indus'],axis=1)
```

```python
model_1 = sma.OLS(y,xc).fit() 
model_1.summary()
```

```python
residuals = model_1.resid
```

```python
sns.kdeplot(residuals)
```

```python
import scipy.stats as stats
```

```python
stats.probplot(residuals,plot=plt)
plt.show()
```

```python
stats.jarque_bera(residuals)
```

For jarque bera

h0 : Data is normal ,
ha : Data is not normal ,
P_value < 0.05. So reject h0. Data is not normal

Log Transformation

```python
new_y = np.log(y)
```

```python
model_2 = sma.OLS(new_y,xc).fit()
#model_2.summary()
```

```python
stats.jarque_bera(model_2.resid)
```

BoxCox Transformation

```python
box_y,lam = stats.boxcox(y)
lam
```

```python
model_3 = sma.OLS(box_y,xc).fit()
#model_3.summary()
```

```python
stats.jarque_bera(model_3.resid)
```

# Assumption - 3 Test of Homoscadasticity

```python
model_4 = sma.OLS(y, xc).fit()
#model_4.summary()
```

```python
residuals = model_4.resid
y_pred = model_4.predict(xc)
plt.figure(figsize=(5,5))
sns.regplot(x=y_pred, y=residuals)
```

h0 : The errors terms are homoscedasticity

ha : The errors terms are heteroscedasticity

```python
import statsmodels.stats.api as sma
sma.het_goldfeldquandt(y=model_4.resid, x=xc)
```

```python
sma.het_breuschpagan(model_4.resid, xc)
```

```python
sns.regplot(x=y_pred, y=residuals, lowess=True, line_kws={'color':'red'})
```

```python
from sklearn.metrics import r2_score, mean_squared_error

y_pred_test = lr.predict(xtest)
r2_test = r2_score(ytest, y_pred_test)
r2_test
```

```python
y_pred_train = lr.predict(xtrain)
r2_train = r2_score(ytrain, y_pred_train)
r2_train
```

Complete Code for Linear Regression

```python
lr = LinearRegression()
lr.fit(xtrain, ytrain)

y_pred_test = lr.predict(xtest)
y_pred_train = lr.predict(xtrain)

r2_test = r2_score(ytest, y_pred_test)
r2_train = r2_score(ytrain, y_pred_train)

rmse_test = np.sqrt(mean_squared_error(ytest, y_pred_test))
rmse_train = np.sqrt(mean_squared_error(ytrain, y_pred_train))

results = [r2_test,r2_train,rmse_test,rmse_train]
results
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
```

FEATURES BACKWARD

```python
sfs_backward = sfs(estimator=lr,
                  k_features=6,
                  scoring='r2',
                  forward=False)
sfs_backward = sfs_backward.fit(xtrain, ytrain)
features_back = list(sfs_backward.k_feature_names_)
features_back
```

```python
lr=LinearRegression()
lr.fit(xtrain[features_back],ytrain)

y_pred_train=lr.predict(xtrain[features_back])
y_pred_test=lr.predict(xtest[features_back])

train_r2=r2_score(ytrain,y_pred_train)
test_r2=r2_score(ytest,y_pred_test)

rmse_train=np.sqrt(mean_squared_error(ytrain,y_pred_train))
rmse_test=np.sqrt(mean_squared_error(ytest,y_pred_test))

res_lr = [train_r2,test_r2,rmse_train,rmse_test]
res_lr
```

```python
res = pd.DataFrame()
```

```python
res['Full_Model'] = results
res['Backward_Elimination'] = res_lr
```

```python
res.index = ['R2_Train','R2_Test','RMSE_Train','RMSE_Test']
```

```python
res
```

# Regularization

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet
```

```python
ridge_model= Ridge()
ridge_model.fit(xtrain,ytrain)

y_pred_train= ridge_model.predict(xtrain)
y_pred_test= ridge_model.predict(xtest)

r2_train = r2_score(ytrain,y_pred_train)
r2_test= r2_score(ytest,y_pred_test)

train_rmse= np.sqrt(mean_squared_error(ytrain, y_pred_train))
test_rmse= np.sqrt(mean_squared_error(ytest, y_pred_test))

res_ridge=  [r2_train,r2_test,train_rmse,test_rmse]
res_ridge
```

GridSearchCV for Ridge...

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {'alpha':[0.1, 0.01, 0.2, 0.4, 0.5]}

r = Ridge()

rcv = GridSearchCV(r,parameter, cv=5, scoring='r2')
rcv.fit(xtrain, ytrain)
rcv.best_params_
```

```python
res['Ridge'] = res_ridge
res
```

```python
lasso_model = Lasso()
lasso_model.fit(xtrain, ytrain)

y_pred_train = lasso_model.predict(xtrain)
y_pred_test = lasso_model.predict(xtest)

r2_train = r2_score(ytrain, y_pred_train)
r2_test = r2_score(ytest, y_pred_test)

rmse_train = np.sqrt(mean_squared_error(ytrain, y_pred_train))
rmse_test = np.sqrt(mean_squared_error(ytest, y_pred_test))

res_lasso = [r2_train, r2_test, rmse_train, rmse_test]
res_lasso
```

GridSearchCV for Lasso...

```python
from sklearn.model_selection import GridSearchCV
parameter = {'alpha':[0.1, 0.01, 0.2, 0.4, 0.5]}

l = Lasso()

rcv = GridSearchCV(l,parameter, cv=5, scoring='r2')
rcv.fit(xtrain, ytrain)
rcv.best_params_
```

```python
res['Lasso'] = res_lasso
res
```
