```python
#BAR PLOT 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
name = ['Tata','Birla','Wipro','MRF']
e1 = [100,200,300,400]
e2 = [150,175,250,450]
```

```python
plt.figure(figsize=(3,2), dpi=100)
plt.bar(x = name, height = e1, color='y')
plt.bar(x = name, height = e2, color='g')
plt.show()
```

```python
plt.figure(figsize=(3,2), dpi=100)
plt.bar(x = name, height = e1, color='y')
plt.bar(x = name, height = e2, bottom = e2, color='b')  #Add bottom to show seperately..
plt.show()
```

```python
plt.figure(figsize=(3,2), dpi=100)
xtick_values = np.array(range(len(name)))
print(xtick_values)

plt.bar(x=xtick_values, height=e1, color='blue', width=0.4)
plt.bar(x=xtick_values+0.4, height=e2, color='red', width=0.4)

plt.xticks(ticks = xtick_values+0.2, labels= name)
plt.show()
```
