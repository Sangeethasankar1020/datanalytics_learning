```python
#ACCESS DATAFRAMES TOPICS..
# GRAMMER
# 1. Direct Access
# 2. Integer Location Based Access
# 3. Label Based Access

# 1. Direct Access

    # df['COL_ONLY']
    # df[[LIST OF COL LABELS ONLY]]
    # df[SLICE OF ROW LABELS ONLY]   # END POINT = TRUE
    
# 2. Integer Location Based Access

    # df.iloc[ROW_INDEX]     ROW is Returned
    # df.iloc[ROW_INDEX, COL_INDEX]  COL is returned
    # df.iloc[ROW_SLICE]   ROW is sliced
    # df.iloc[ROW_SLICE, COL_SLICE]  COl is sliced
    # df.iloc[[list of ROW]] #Brings all ROWS
    # df.iloc[[list of ROW], [list of COL]]
    #Combo is also possible
    
# 3. Label Based Access

    #ILOC for numerical slice
    #LOC for string mentional slice
```

```python
def sales_performance(sales):
    if sales > 950:
        return 'Excellent'
    elif sales < 930:
        return 'Average'
    else:
        return 'Good'
```

```python
df7['Sales_Performance'] = df7.Vol.map(sales_performance)
```
