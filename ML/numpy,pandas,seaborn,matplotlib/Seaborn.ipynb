# Seaborn

* Seaborn is used for Data Visualization.
* It is based on Matplotlib.
* It provides a High-Level Interface for drawing attractive and informative statistical graphics.

## Functionalities:
* Allows comparison between multiple variables(Columns/Features).
* Supports Multi-Plot Grids.
* Univariate and Bivariate visualization.
* Availability of different color palettes.
* Estimates and Plots Linear Regression line.

# Importing Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.isnull().sum()
```

```python
sns.boxplot(df)
```

```python
df.shape
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#IQR
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3-q1

ll = q1 - 1.5*iqr
ul = q3 + 1.5*iqr
print(ll)
print("-------------------")
print(ul)
```

```python
num = df.select_dtypes(include='number').columns
num
```

```python
outliers = df[((df[num]<ll) | (df[num]>ul)).any(axis=1)]
outliers
```

```python
outliers.shape
```

# Strip Plot

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.stripplot(x = 'sex', y = 'age', data = df,color='g') #cat v/s num
plt.show()
```

# cat vs num and cat

```python
sns.stripplot(x = 'sex', y = 'age', hue = 'survived', data = df)
plt.show()
```

# Regression plot

* A regression plot is used to visualize the relationship between two continuous variables along with a fitted regression model. It is particularly useful for identifying trends, correlations, and patterns in data.

```python
df = sns.load_dataset('iris')
df.head()
```

```python
sns.regplot(x = df['sepal_length'], y = df['petal_length'])
```

# Joint Plot
* A joint plot in Seaborn provides a way to visualize the relationship between two variables, along with their univariate distributions (histograms or density plots) on the margins. It is a combination of a scatter plot (or other bivariate plot) and univariate plots for each variable, which helps in understanding the correlation and individual distributions in a single visualization.

```python
sns.jointplot(x = df['age'], y=df['fare'])
```
