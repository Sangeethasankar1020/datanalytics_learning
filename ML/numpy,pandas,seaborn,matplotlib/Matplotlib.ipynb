**Types of Plots**:
- *Line Plot*: Shows data points connected by straight lines.
- *Scatter Plot*: Displays individual data points on a graph to show relationships.
- *Bar Plot*: Represents data with rectangular bars to compare categories.
- *Histogram*: Displays the distribution of numerical data using bars.
- *Box Plot*: Shows the distribution of data along with statistical information.
- *Pie Chart*: Represents data in a circular graph divided into slices to show proportions.

**Comparing Column Types and Recommended Plots**:
- **One Variable**:
  - *For a single column of numerical data*: Histogram or Box Plot for distribution.
  - *For a single column of categorical data*: Bar Plot or Pie Chart for category frequencies.

- **Two Variables**:
  - *For comparing two numerical columns*: Scatter Plot for relationship visualization.
  - *For one categorical and one numerical column*: Bar Plot or Box Plot for distribution.

- **Three Variables**:
  - *For comparing three numerical variables*: 3D Scatter Plot or Bubble Chart for relationships.

- **Multiple Variables**:
  - *For comparing multiple variables*: Multiple Line Plots, Stacked Bar Plots, or Grouped Bar Plots for effective display. Heatmaps for correlations.

Choose the plot type based on your data characteristics and the insights you aim to derive.

# Bar Plot - Cat vs Num & Cat alone

```python
# Create a List of bill amounts
amount = [3000, 1200, 5000, 1800, 4500]
customer = ('John', 'Ross', 'Rick', 'Mia', 'Dima')

# x- categorical variable
# Create a bar plot
plt.bar(x=customer, height=amount)#x - cat, height - num

# Add axes and plot labels
plt.title('Distribution of Bill Amount')
plt.xlabel('Customer')
plt.ylabel('Bill Amount')

# Display the plot
plt.show()
```

#  Horizontal bar plot

```python
plt.barh(y = customer, width = amount) #y-cat, width-num
plt.show()
```

# Grouped Bar Plot


```python
# Create the data for marks of 5 students
Python_marks = (50, 65, 40, 35, 77)
R_marks = (55, 72, 94, 70, 85)

# Set the position of bars
index = np.arange(5)

# Plot a bar plot for each subject
# 'x' represents position of bar
# 'height' represents value of the bar
# 'width' represents width of the bar
# 'Label' assigns Label to the bar

plt.bar(x =index, height=Python_marks, width = 0.35, label='Python')
plt.bar(x =index+0.35, height=R_marks, width = 0.35, label = 'R')
# Add axes and plot labels
plt.xlabel('Students')
plt.ylabel('Scores')
plt.title('Scores by Students')

# 'ticks' assigns position of Label
# 'Labels' assigns Label to each bar
# Plot the bar plot for each subject
plt.xticks(ticks=index + 0.35 / 2, labels=('A','B','C','D', 'E'))

# Add the Legend
plt.legend()

# Display the plot
plt.show()
```

# Stacked Bar Plot

```python
# Create the data for marks of 5 students
Python_marks = (50, 65, 40, 35, 77)
R_marks = (75, 72, 64, 60, 85)

# Set the position of bars
index = np.arange(5)

# Plot a stacked bar plot for each subject
# 'x' represents position of bar
# 'height' represents value of the bar
# 'bottom' represents the bar plot at bottom
# 'Label' assigns label to the bar
plt.bar(x=index, height=Python_marks, label='Python')
plt.bar(x=index, height=R_marks, bottom=Python_marks, label='R')

# Add axes and plot labels
plt.xlabel('Students')
plt.ylabel('Scores')
plt.title('Scores by Students')

# 'ticks' assigns position of Label
# Plot the 'R_marks' above the 'Python_marks'
# 'Labels' assigns Label to each bar
plt.xticks(ticks=index, labels=('A', 'B', 'C', 'D', 'E'))

# Add the Legend
plt.legend()

# Display the plot
plt.show()
```

# Pie Plot

```python
# Create a List of bill amounts
amount = [3000, 1200, 5000, 1800, 4500]
customer = ('John', 'Ross', 'Rick', 'Mia', 'Dima')

# Position of bars
x_pos = np.arange(len(customer))  # 'x' represents categorical variable

# Create a vertical bar plot
plt.bar(x=x_pos, height=amount)

# Add labels to each bar
plt.xticks(x_pos, customer)

# Add axes and plot labels
plt.title('Distribution of Bill Amount')
plt.xlabel('Customer')
plt.ylabel('Bill Amount')

# Display the plot
plt.show()
```
