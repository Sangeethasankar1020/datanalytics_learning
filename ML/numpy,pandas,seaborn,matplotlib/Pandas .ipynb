# Mapping and Replacing

```python
df_items = pd.DataFrame ({'Product' : ['Milk' , 'Cornflakes' , 'Prunes' , 'Bread' , 'Jam'], 'Price': [15, 134, 322, 16, 165]} ) 
df_items
```

```python
brand = {'Milk':'Milkman', 'Prunes':'DryFruits',
        'SauSage':'ColdChiken','Cornflakes':'Kellogs',
        'Jam':'DailyEats','Bread':'Bakes&More'}
#map the dictionary to 'df_items'
df_items['Brand'] = df_items['Product'].map(brand)
```

```python
price = {15:30,322:324}
#replace the price 15 by 30 and 322 by 324
df_items['Price'].replace(price, inplace=True)
df_items
```
