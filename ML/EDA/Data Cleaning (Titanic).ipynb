```python
q1 = df[num].quantile(0.25)
q3 = df[num].quantile(0.75)
iqr = q3 - q1
ll = q1 - 1.5*iqr
ul = q3 + 1.5*iqr
ll,ul
```

```python
outliers = df[((df[num]<ll) | (df[num]>ul)).any(axis=1)]
outliers
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
df.skew()
```

```python
#Scaling - -is used to make the magnitude of all the numerical variables same
ss = StandardScaler()#for normally distributed  -1 to +1
rs = RobustScaler()#if had outliers 
ms = MinMaxScaler()#when data is skewed - 0 to 1
df['age_ss'] = ss.fit_transform(df[['age']])
df['fare_ms'] = ms.fit_transform(df[['fare']])
```
