```python
mu,sigma=0,0.5
```

```python
s=np.random.normal(mu,sigma,1000)
```

```python
plt.hist(s,bins=30,density=True,alpha=0.5,color='g')
```

```python
xmin,xmax=plt.xlim()
x=np.linspace(xmin,xmax,100)
p=stats.norm.pdf(x,mu,sigma)
plt.plot(x,p,'k',linewidth=1)
plt.title('Normal Distribution')
plt.show()
```
