```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as  np
```

```python
x=np.linspace(1,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Example")
plt.show()
```

```python
#bar
```

```python
x=np.array(["A", "B", "C", "D"])
y=np.array([3,6,9,12])

plt.bar(x,y)
plt.xlabel("Categorical Value")
plt.ylabel("Numerical Value")
plt.title("Bar Chart")
plt.show()
```

```python
#subplot
```

```python
x=np.linspace(1,5,11)
y=x**2
plt.subplot(1,2,1)
plt.plot(x,y,"g--")
plt.subplot(1,2,2)
plt.plot(y,x,"r*-")
plt.show()
```
