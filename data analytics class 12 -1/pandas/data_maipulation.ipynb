# Data Manipulation 

```python
import pandas as pd
```

```python
data = {'Product': ['Laptop', 'Mobile', 'Tablet', 'Monitor'],
        'Price': [1200, 800, 300, 400],
        'Units_Sold': [100, 150, 80, 60]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
#new column
```

```python
df=df.assign(Total_Revenue=df["Price"]*df["Units_Sold"])
```

```python
df
```

```python
#remove column
```

```python
df=df.drop(columns=["Units_Sold"])
```

```python
df
```

```python
#rename
```

```python
df=df.rename(columns={"Price":"Unit_Price","Total_Revenue":"Revenue"})
```

```python
df
```

```python
#rename index
```

```python
df=df.rename(index={0:"A",1:"B",2:"C",3:"D"})
```

```python
df
```

```python
#sorting a data by revenue
```

```python
df=df.sort_values(by="Revenue",ascending=True)
```

```python
df
```

```python
#sort by index
```

```python
df=df.sort_index()
```

```python
df
```

```python
#filtering data
```

```python
filtered_data=df[df["Revenue"]>30000]
```

```python
filtered_data
```

```python
#multiple conditions
```

```python
filtered_data2=df[(df["Unit_Price"]<500) & (df["Revenue"]>20000)]
```

```python
filtered_data2
```

# Merging , Joining, Concate

```python
df1=pd.DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']},index=[0,1,2,3]
)
```

```python
df2=pd.DataFrame({
    'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']},index=[4,5,6,7]
)
```

```python
df3=pd.DataFrame({
    'A':['A8','A9','A10','A11'],
    'B':['B8','B9','B10','B11'],
    'C':['C8','C9','C10','C11'],
    'D':['D8','D9','D10','D11']},index=[8,9,10,11]
)
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
df=pd.concat([df1,df2,df3])
```

```python
df
```

```python
df=pd.concat([df1,df2,df3],axis=1)
```

```python
df
```

```python
#merge
```

```python
left=pd.DataFrame({'key':['K0','K1','K2','K3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']
                  })
```

```python
right=pd.DataFrame({'key':['K0','K1','K2','K3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']
                  })
```

```python
pd.merge(left,right,how="inner",on="key")
```

```python
left=pd.DataFrame({'key1':['K0','K0','K1','K2'],
                   'key2':['K0','K1','K0','K1'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']
                  })
```

```python
right=pd.DataFrame({'key1':['K0','K1','K1','K2'],
                    'key2':['K0','K0','K0','K0'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']
                  })
```

```python
pd.merge(left,right,on=["key1","key2"])
```

```python
pd.merge(left,right,how="outer",on=["key1","key2"])
```
