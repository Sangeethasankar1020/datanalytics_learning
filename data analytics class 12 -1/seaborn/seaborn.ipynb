```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
#data loading
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

### Univariate

```python
#dist plt - distribution plot
```

```python
sns.displot(df['total_bill'],kde=True)
```

```python
#rug plot
```

```python
sns.rugplot(df['total_bill'])
```

```python
#kde plot
```

```python
sns.kdeplot(df['total_bill'])
sns.rugplot(df['total_bill'])
```

### Bivariate 

```python
# joint plot
```

```python
sns.jointplot(x=df['total_bill'],y=df['tip'],data=df)
```

```python
sns.jointplot(x=df['total_bill'],y=df['tip'],data=df,kind='hex')
```

```python
sns.jointplot(x=df['total_bill'],y=df['tip'],data=df,kind='reg')
```

### Multivariate

```python
#pairplot
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex',palette="coolwarm")
```
