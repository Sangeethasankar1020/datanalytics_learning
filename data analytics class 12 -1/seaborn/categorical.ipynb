# Categorical Plot

### Univariate 

```python
import seaborn as sns
import numpy as np
```

```python
data=sns.load_dataset('tips')
```

```python
data
```

```python
sns.barplot(x='sex',y='total_bill',data=data)
```

```python
sns.barplot(x='sex',y='total_bill',data=data,estimator=np.std)
```

```python
#countplot
```

```python
sns.countplot(x='sex',data=data)
```

```python
#boxplot
```

```python
sns.boxplot(x='sex',y='total_bill',data=data,palette="rainbow")
```

```python
sns.boxplot(data=data,palette="rainbow")
```

```python
sns.boxplot(data=data,palette="rainbow",orient='h')
```

```python
sns.boxplot(x='sex',y='total_bill',data=data,hue='smoker',palette="rainbow")
```

### Bivariate 

```python
#violin plot
```

```python
sns.violinplot(x="tip", y="day", data=data,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=data,color='black',size=3)
```
