```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Duration': [30, 45, 50, 60, 30, 40, 70, 80, 20, 50],
    'Calories': [150, 250, 300, 400, 200, 220, 450, 500, 130, 280],
    'Maxpulse': [110, 150, 130, 175, 140, 120, 160, 190, 100, 140]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Plot a line graph
df.plot()
plt.title('Workout Duration and Calories Over Time')
plt.xlabel('Workout Session')
plt.ylabel('Value')
plt.show()
```

###  Scatter plot

```python
#A scatter plot is useful for showing relationships between two variables. Let's plot Duration vs. 
#Calories to see how workout duration affects calories burned.
```

```python
df.plot(kind='scatter', x='Duration', y='Calories', color='blue')
plt.show()
```

```python
plt.title('Calories Burned vs Workout Duration')
```

```python
plt.xlabel('Duration (minutes)')
```

```python
plt.ylabel('Calories Burned')
```

```python
# Scatter plot for Duration vs Maxpulse
df.plot(kind='scatter', x='Duration', y='Maxpulse', color='red')
plt.title('Maxpulse vs Workout Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Maxpulse (bpm)')
plt.show()


```

```python
# Scatter plot for Duration vs Calories
df.plot(kind='scatter', x='Duration', y='Calories', color='blue')
plt.title('Calories Burned vs Workout Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Calories Burned')
plt.show()
```

### Histogram 

```python
#A histogram shows the frequency distribution of a column. For example,
#let's see how many workouts lasted between specific intervals of time:
```

```python
# Histogram for Duration
df['Duration'].plot(kind='hist', bins=5, color='green')
plt.title('Distribution of Workout Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
#multiple histogram
```

```python
#alpha=0.5: Makes the histograms slightly transparent, so you can see both distributions.
```

```python
# Multiple histograms
df[['Duration', 'Calories']].plot(kind='hist', bins=5, alpha=0.5)
plt.title('Distribution of Workout Duration and Calories Burned')
plt.show()
```
