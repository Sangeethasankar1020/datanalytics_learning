### pairplot 

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns).



```python
import seaborn as sns
%matplotlib inline
```

```python
df=sns.load_dataset('tips')
df
```

```python
sns.pairplot(df)
#efficient for small data
#time does not comes under categorical also day not comes under actegorical
```

```python
sns.pairplot(df, hue='sex', palette='coolwarm') #palette design
```

In order for a heatmap to work properly, your data should already be in a matrix form, the sns.heatmap function basically just colors it in for you. For example:



```python
#selecting numberic only for correlation
```

```python
numeric_df=df.select_dtypes(include='number')
```

```python
numeric_df.corr()
```

```python
#its as positive co-relation - above .5 is good corellation , we have to ask domain experise also 
```

```python
#heatmap - for correlation
```

```python
sns.heatmap(numeric_df.corr())
```

```python
#value view
sns.heatmap(numeric_df.corr(),annot=True,cmap='coolwarm')
```
