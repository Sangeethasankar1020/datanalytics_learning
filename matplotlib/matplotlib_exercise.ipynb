# Matplotlib exercise 

### Data 

```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('Y')
axes.set_title('title')
```

### 2 

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,.2,.2])
```

### 3

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,y,'r')
axes2=fig.add_axes([0.2,0.5,.2,.2])
axes2.plot(x,y,'r')
```

### 5 

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,.4,.4])

axes1.plot(x,z,'b')
axes1.set_xlabel('x')
axes1.set_ylabel('N')
axes1.set_xlim([0,100])
axes1.set_ylim([0,10000])



axes2.plot(x,y,'b')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoom')
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
```
