# Group by

Groupby is a powerfull method used to group data based on one or more colmns and perform aggregrate operations on each group

data aggregration:

  
    

  -  mean()
  -  median()
  -  max()
  -  min()
  -  count()
  -  sum()

for single column

grouped_df=df.groupby('column name').sum()

for multiple columns

grouped_df =df.groupby(['col1','col2'])['numerical_col2'].mean()

```python
#single columns
grouped_df=df.groupby('Branch')
grouped_count=grouped_df['Branch'].count()
grouped_count
```

# Data visulazition

Data visulization aims to represnt data in a visual context to make it easier to understand hightlights important aspects and communicates findings effecttively to stakeolders

 - Bar chart
 - Line chart
 - scatter
 - Histogram

```python
#barplot
```

```python
plt.figure(figsize=(8,6)) #plot size
sns.barplot(x='Gender',y='gross margin percentage',data=df,color='pink',edgecolor='red')
plt.title('Percentage vs Gender')
plt.xlabel("Gender")
plt.ylabel('Gross margin percentage')
plt.show()
```

```python
#histogram
#gross income data distribution
```

```python
plt.figure(figsize=(8,6))
#to view accurate apply kde
sns.histplot(df['gross income'],bins=30,color="teal",kde=True)
plt.show()
```

```python
#scatter plot
#two numerical columns
```

```python
sns.scatterplot(x='Quantity',y='gross income',data=df,hue=df['Gender'])
plt.title("Relationship between quantity vs gross income")
plt.show()
```

```python
#line plot
#continuos data plot
```

```python
sns.lineplot(x="Invoice ID",y="gross income",data=df,marker='o',ls='-.')
```
