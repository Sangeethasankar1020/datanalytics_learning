# Data Visualization

Data visualization is a critical aspect of exploratory data analysis (EDA) in data analytics. It helps analysts understand data better and communicate findings effectively

###  univariate

Univariate analysis focuses on a single variable. The goal is to understand its distribution, central tendency, and spread.

 - histograms
 - box plots
 - bar plots

```python
import pandas as pd

data = {'Age': [23, 45, 31, 34, 26, 29, 55, 60, 41, 38]}
df = pd.DataFrame(data)
```

```python
# Summary statistics
summary = df.describe()
print(summary)
```

```python
#visualize
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=5, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### bivariate

Bivariate analysis examines the relationship between two variables. The goal is to understand how they correlate or influence each other.

 - scatter plot
 - bar chart (grouped)
 - heatmap for correlation matrices

```python
#correaltion
```

```python
data_bivariate = {
    'Age': [23, 45, 31, 34, 26, 29, 55, 60, 41, 38],
    'Income': [30000, 60000, 45000, 48000, 32000, 35000, 80000, 90000, 70000, 52000]
}
df=pd.DataFrame(data_bivariate)
```

Use the Pearson correlation coefficient to assess the strength and direction of the relationship.

```python
correlation = df.corr()
print(correlation)
```

```python
#visulaize
```

```python
plt.scatter(df['Age'], df['Income'])
plt.title('Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

###  multivariate 

Multivariate analysis involves three or more variables. The goal is to understand complex relationships between them.

- pair plot
- 3D scatter plots
- heatmap
- faceted plt
  

```python
data_multivariate = {
    'Age': [23, 45, 31, 34, 26, 29, 55, 60, 41, 38],
    'Income': [30000, 60000, 45000, 48000, 32000, 35000, 80000, 90000, 70000, 52000],
    'Education Level': [1, 3, 2, 2, 1, 1, 3, 4, 3, 2]  # 1: High School, 2: Bachelor's, 3: Master's, 4: PhD
}
df=pd.DataFrame(data_multivariate)
```

```python
df
```

```python
df.corr()
```

```python
import seaborn as sns

sns.pairplot(df)
plt.show()
```
